```python
import pandas as pd
```

```python
df = pd.read_csv('./trip.csv', delimiter=',', error_bad_lines=False)
df
```

```python
#the average trip duration for a borrowed bicycle
avg_trip = df['tripduration'].median()
avg_trip
```

```python
# the most common age of a bicycle-sharer
birth_year = df['birthyear'].value_counts().idxmax()
current_year = pd.datetime.now().year
age = int(current_year - birth_year)
age
```
